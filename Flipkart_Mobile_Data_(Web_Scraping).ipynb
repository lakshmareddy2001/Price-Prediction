```python
names=[]
price=[]
rating=[]
ram=[]
memory=[]
battery=[]
camera=[]
screen=[]
for k in range(1,42):
  url="https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_2_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_2_0_na_na_na&as-pos=2&as-type=TRENDING&suggestionId=mobiles&requestId=04978a72-effe-498e-b80f-6931338bc3d5&page="+str(k)
  r=requests.get(url)
  htmlContent=r.content
  soup=BeautifulSoup(htmlContent,'html.parser')
  for a in soup.find_all('a',class_="_1fQZEK"):
    t=[]
    for i in a.find_all("li",class_='rgWa7D'):
      t.append(i.get_text())
    memory.append(t[0])
    screen.append(t[1])
    camera.append(t[2])
    battery.append(t[3])
    try:
      names.append(a.find('div',class_="_4rR01T").get_text())
    except:
      names.append("Nan")
    price.append(a.find('div',class_="_30jeq3 _1_WHN1").get_text())
    try:
      rating.append(a.find('div',class_="_3LWZlK").get_text())
    except:
      rating.append("Nan")

```

```python
len(names)
```

```python
data={"Name":names,"Price":price,"Rating":rating,"Memory":memory,"Battery":battery,"Camera":camera,"Screen":screen}
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data)
```
